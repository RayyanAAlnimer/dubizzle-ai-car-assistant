# Screenshot 1
## Complete Multi-turn Booking and Lead Flow

```python
import httpx

API_URL = "http://127.0.0.1:8000/chat"


def chat(user_id, message):
    response = httpx.post(
        API_URL,
        json={
            "user_id": user_id,
            "message": message,
        },
        timeout=30,
    )
    response.raise_for_status()

    reply = response.json()["reply"]
    print(f"User ID: {user_id}")
    print(f"User: {message}")
    print(f"Assistant: {reply}")
    print("-" * 80)

    return reply
```

```python
demo_user = "demo_user"

chat(demo_user, "Find cars with warranty and GCC specs")
chat(demo_user, "Tell me about the second one")
chat(demo_user, "Book the second one")
chat(demo_user, "Tuesday")
chat(demo_user, "4 PM")
chat(demo_user, "150000")
chat(demo_user, "Family")
chat(demo_user, "0500000000")
```
